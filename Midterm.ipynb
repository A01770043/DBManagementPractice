```python
# Import sqlite3 Library
import sqlite3

# Connect to SQLite DataBase
conn = sqlite3.connect('company_database.db')
print("Database Open Succesfully")

#Create Vehicle table
conn.execute('''
CREATE TABLE IF NOT EXISTS Vehicle (
    Vehicle_ID VARCHAR(25) PRIMARY KEY NOT NULL,
    Registration_Number VARCHAR(25) NOT NULL,
    Make VARCHAR(25) NOT NULL,
    Model VARCHAR(25) NOT NULL,
    Color VARCHAR(25) NOT NULL,
    Year INT,
    Current_Odometer INT NOT NULL,
    Passenger_Capacity INT NOT NULL
);
''')

#Insert example data into Vehicle table
conn.execute("INSERT INTO Vehicle (Vehicle_ID, Registration_Number, Make, Model, Color, Year, Current_Odometer, Passenger_Capacity) VALUES ('V1000', '2001 ABC', 'Volvo', 'XC90 SE', 'Silver', 2001, 4350, 5)")
conn.execute("INSERT INTO Vehicle (Vehicle_ID, Registration_Number, Make, Model, Color, Year, Current_Odometer, Passenger_Capacity) VALUES ('V1001', '4350 AFD', 'Kia', 'K7', 'Black', 2006, 2195, 4)")
conn.execute("INSERT INTO Vehicle (Vehicle_ID, Registration_Number, Make, Model, Color, Year, Current_Odometer, Passenger_Capacity) VALUES ('V1002', '2195 AHR', 'Tesla', '2020 F', 'White', 2021, 509, 4)")
conn.execute("INSERT INTO Vehicle (Vehicle_ID, Registration_Number, Make, Model, Color, Year, Current_Odometer, Passenger_Capacity) VALUES ('V1003', '509 DXF', 'Ford', 'Transit', 'Silver', 2020, 974, 2)")

#Print Vehicle table
cursor = conn.execute("SELECT Vehicle_ID, Registration_Number, Make, Model, Color, Year, Current_Odometer, Passenger_Capacity FROM Vehicle")
print("Data from the Vehicle table:")
for row in cursor:
    print("Vehicle ID =", row[0])
    print("Registration Number =", row[1])
    print("Make =", row[2])
    print("Model =", row[3])
    print("Color =", row[4])
    print("Year =", row[5])
    print("Current Odometer =", row[6])
    print("Passenger Capacity =", row[7])
    print()
```

```python
# Create Maintenance_Repair_Action table
conn.execute('''
CREATE TABLE IF NOT EXISTS Maintenance_Repair_Action (
    Maintenance_Repair_Action_ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Vehicle_ID VARCHAR(25) NOT NULL,
    Action_Type CHAR(1) NOT NULL,
    Odometer INT NOT NULL,
    Cost DECIMAL(10, 2) NOT NULL,
    Description VARCHAR(255) NOT NULL,
    Date DATE NOT NULL,
    FOREIGN KEY (Vehicle_ID) REFERENCES Vehicle(Vehicle_ID)
);
''')

# Insert example data into Maintenance_Repair_Action table
conn.execute("INSERT INTO Maintenance_Repair_Action (Vehicle_ID, Action_Type, Odometer, Cost, Description, Date) VALUES ('V1000', 'M', 5000, 200.00, 'Routine maintenance', '2022-05-10')")
conn.execute("INSERT INTO Maintenance_Repair_Action (Vehicle_ID, Action_Type, Odometer, Cost, Description, Date) VALUES ('V1001', 'R', 22000, 800.00, 'Engine repair', '2023-02-15')")
conn.execute("INSERT INTO Maintenance_Repair_Action (Vehicle_ID, Action_Type, Odometer, Cost, Description, Date) VALUES ('V1003', 'M', 10000, 150.00, 'Brake replacement', '2023-01-20')")

# Print Maintenance_Repair_Action table
cursor = conn.execute("SELECT Maintenance_Repair_Action_ID, Vehicle_ID, Action_Type, Odometer, Cost, Description, Date FROM Maintenance_Repair_Action")
print("Data from the Maintenance/Repair Actions table:")
for row in cursor:
    print("Maintenance/Repair Action ID =", row[0])
    print("Vehicle ID =", row[1])
    print("Action Type =", row[2])
    print("Odometer =", row[3])
    print("Cost =", row[4])
    print("Description =", row[5])
    print("Date =", row[6])
    print()
```

```python
# Create Country table
conn.execute('''
CREATE TABLE IF NOT EXISTS Country (
    Geographical_Code VARCHAR(2) PRIMARY KEY NOT NULL,
    Spoken_Language VARCHAR(100),
    Name VARCHAR(100)
);
''')

# Insert example data into Country table
conn.execute("INSERT INTO Country (Geographical_Code, Name, Spoken_Language) VALUES ('KR', 'South Korea', 'Korean, English')")
conn.execute("INSERT INTO Country (Geographical_Code, Name, Spoken_Language) VALUES ('US', 'United States', 'English, Spanish')")
conn.execute("INSERT INTO Country (Geographical_Code, Name, Spoken_Language) VALUES ('JP', 'Japan', 'Japanese')")
conn.execute("INSERT INTO Country (Geographical_Code, Name, Spoken_Language) VALUES ('CN', 'China', 'Mandarin')")

# Print Country table
cursor = conn.execute("SELECT Geographical_Code, Spoken_Language, Name FROM Country")
print("Data from the Country table:")
for row in cursor:
    print("Geographical Code =", row[0])
    print("Spoken Language =", row[1])
    print("Name =", row[2])
    print()
```
